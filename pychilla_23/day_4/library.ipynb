```python
import pandas as pd
# pd.read_excel()
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# plotting
import seaborn as sns
sns.set_theme(style="whitegrid")
# Load the diamonds dataset from seaborn
# This dataset contains information about diamonds, including their carat, cut, color, clarity, and price.
diamonds = sns.load_dataset("diamonds")
# Display the first few rows of the dataset
clarity_ranking = ["IF", "VVS1", "VVS2", "VS1", "VS2", "SI1", "SI2", "I1"]
sns.boxenplot(x="clarity", y="price", width_method="linear", data=diamonds, order=clarity_ranking)
```

```python
import plotly.express as px
# Create a box plot using Plotly Express
df = px.data.gapminder()
px.scatter(df, x="gdpPercap", y="lifeExp", color="continent", size="pop",
           hover_name="country", log_x=True, size_max=60,
           animation_frame="year", range_x=[100, 100000], range_y=[25, 90],
           title="GDP per Capita vs Life Expectancy by Continent")
```
